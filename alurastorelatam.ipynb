# **Importación de datos**



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
total_tienda = tienda['Precio'].sum()
total_tienda2 = tienda2['Precio'].sum()
total_tienda3 = tienda3['Precio'].sum()
total_tienda4 = tienda4['Precio'].sum()
```

```python
print("El total de ventas en la tienda 1 es de " + str(total_tienda))
print("El total de ventas en la tienda 2 es de " + str(total_tienda2))
print("El total de ventas en la tienda 3 es de " + str(total_tienda3))
print("El total de ventas en la tienda 4 es de " + str(total_tienda4))
```

# 2. Ventas por categoría

## **Conteo de ventas por categoria Tienda 1**

```python
conteo_categoria1 = tienda.groupby('Categoría del Producto').size().reset_index(name='Cantidad').sort_values(by='Cantidad',ascending=False)
conteo_categoria1
```

#6. Graficos

## **Grafica de barras del cantidad vendida por categoría**

```python
import matplotlib.pyplot as plt
import numpy as np

categorias = conteo_categoria1['Categoría del Producto']
x = np.arange(len(categorias))
width = 0.2

plt.figure(figsize=(12,6))

plt.bar(x, conteo_categoria1['Cantidad'], width)
plt.bar(x + width, conteo_categoria2['Cantidad'], width)
plt.bar(x + 2*width, conteo_categoria3['Cantidad'], width)
plt.bar(x + 3*width, conteo_categoria4['Cantidad'], width)

plt.xticks(x + 1.5*width, categorias, rotation=45)
plt.xlabel('Categoría')
plt.ylabel('Cantidad de productos vendidos')
plt.title('Ventas por categoría por tienda')
plt.legend(['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'])

plt.tight_layout()
plt.show()
```

## **Grafico pastel de total vendido por tienda**

```python
from matplotlib.ticker import FuncFormatter

plt.plot(
    ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],             # eje X
    [total_tienda, total_tienda2, total_tienda3, total_tienda4],  # eje Y
    marker='o'  # puntos visibles
)

plt.xlabel('Tiendas')
plt.ylabel('Valor vendido')
plt.title('Total vendido por tienda')
plt.gca().yaxis.set_major_formatter( FuncFormatter(lambda x, p: f'{x/1e6:.0f}M') )

plt.grid(True)
plt.show()
```

## **Grafica de barras calificación promedio por tienda**

```python
plt.figure(figsize=(8,5))

plt.barh(
    ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],   # categorías
    [calificacion_prom1, calificacion_prom2, calificacion_prom3, calificacion_prom4]    # valores numéricos
)

plt.xlabel('Promedio de calificación')
plt.ylabel('Tiendas')
plt.title('Calificaciíon promedio por tienda')
plt.gca().invert_yaxis()

plt.tight_layout()
plt.show()
```

# **Conclusión**

Con base en el análisis de los datos, se identifica que la Tienda 4 es la menos rentable y por lo tanto, la principal candidata para su venta. Esta tienda presenta el menor valor total vendido en comparación con las demás, lo que evidencia un desempeño económico inferior.



```python
total_ventas_tiendas = pd.DataFrame({
    'Tienda' : ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
    'Valor Total Vendido' : [int(total_tienda), int(total_tienda2), int(total_tienda3), int(total_tienda4)]
}).set_index('Tienda')
total_ventas_tiendas.sort_values('Valor Total Vendido')
```

Adicionalmente, su volumen de ventas por categoría es consistentemente más bajo, sin destacar significativamente en ninguna categoría específica.

Aunque la calificación promedio de los clientes es similar al resto de las tiendas, este factor no compensa el bajo rendimiento en ingresos.

```python
costo_envio_promedio_tiendas = pd.DataFrame({
    'Tienda' : ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
    'Calificación Promedio' : [ calificacion_prom1, calificacion_prom2, calificacion_prom3, calificacion_prom4 ]
}).set_index('Tienda')
costo_envio_promedio_tiendas.sort_values('Calificación Promedio')
```


En consecuencia, desde una perspectiva financiera y estratégica, la Tienda 4 representa la opción menos rentable y la más adecuada para ser vendida.
